```python
def extract_face(filename, required_size=(160, 160)):
    # load image from file
    image = Image.open(filename)
    # convert to RGB, if needed
    image = image.convert('RGB')
    # convert to array
    pixels = np.asarray(image)
    # create the detector, using default weights
    detector = MTCNN()
    # detect faces in the image
    results = detector.detect_faces(pixels)
    # extract the bounding box from the first face
    x1, y1, width, height = results[0]['box']
    # deal with negative pixel index
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 + height
    # extract the face
    face = pixels[y1:y2, x1:x2]
    # resize pixels to the model size
    image = Image.fromarray(face)
    image = image.resize(required_size)
    face_array = np.asarray(image)
    return face_array
```

```python
import cv2
import mediapipe as mp

def mediapipe_face_landmarks(image):
    """
    Processes a face image, detects landmarks, and returns the annotated image
    along with a list of landmark positions.

    Args:
        image_path (str): Path to the input face image.

    Returns:
        annotated_image (numpy.ndarray): Image with annotated landmarks.
        landmarks_list (list): A list of (x, y, z) tuples for each detected landmark.
    """
    # Initialize MediaPipe Face Mesh and Drawing utilities
    height, width,_ = image.shape
    mp_face_mesh = mp.solutions.face_mesh
    mp_drawing = mp.solutions.drawing_utils
    mp_drawing_styles = mp.solutions.drawing_styles

    # Load the input image
    # image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image at path '{image_path}' not found.")
    
    # Convert the image from BGR to RGB
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Initialize FaceMesh
    with mp_face_mesh.FaceMesh(
        static_image_mode=True,
        max_num_faces=1,
        refine_landmarks=True,
        min_detection_confidence=0.5) as face_mesh:
        
        # Process the image to detect face landmarks
        results = face_mesh.process(rgb_image)

        # If no landmarks are detected, return the original image and an empty list
        if not results.multi_face_landmarks:
            return image, []

        # Prepare annotated image
        annotated_image = image.copy()
        landmarks_list = []

        # Draw landmarks and extract positions
        for face_landmarks in results.multi_face_landmarks:
            # Draw the face landmarks on the image
            mp_drawing.draw_landmarks(
                image=annotated_image,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_TESSELATION,
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles
                .get_default_face_mesh_tesselation_style())
            
            # Extract the (x, y, z) positions of each landmark
            for landmark in face_landmarks.landmark:
                landmarks_list.append((landmark.x*width, landmark.y*height, landmark.z))

        return annotated_image, landmarks_list
```

```python
import dlib
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")  # Ensure this file is downloaded

def detect_facial_landmarks_dlib(image_path):
    """
    Detects facial landmarks in an image and returns the annotated image along with landmark positions.

    Args:
        image_path (str): Path to the input image.

    Returns:
        image (numpy.ndarray): Image with landmarks drawn.
        landmarks_list (list of dict): A list of dictionaries, one for each detected face, 
                                       containing landmark positions {landmark_idx: (x, y)}.
    """
    # Load image
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError("Image not found.")

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = detector(gray)
    print(faces)
    if not faces:
        print("No face detected.")
        return image, []

    # Prepare list to store landmarks for all detected faces
    landmarks_list = []

    # Draw landmarks for each face
    for face in faces:
        landmarks = predictor(gray, face)
        face_landmarks = {}  # Dictionary to store landmarks for the current face
        for n in range(68):  # 68 landmarks
            x, y = landmarks.part(n).x, landmarks.part(n).y
            face_landmarks[n] = (x, y)
            cv2.circle(image, (x, y), 2, (0, 255, 0), -1)  # Draw landmark on image

        landmarks_list.append(face_landmarks)

    return image, landmarks_list
```
